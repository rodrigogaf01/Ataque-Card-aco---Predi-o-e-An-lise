## Predição e Análise de Ataque Cardíaco

---

O infarto está entre as principais causas de morte no Brasil e no mundo e, por ano, mais de 100 mil brasileiros morrem em decorrência dele, o que equivale a dois estádios do Arena Fonte Nova, em Salvador, lotados em sua capacidade.

Mais conhecido como ataque cardíaco, acontece quando o fluxo do sangue para o coração é interrompido e o coração não recebe sangue e oxigênio suficientes para se manter em atividade

Com isso, o músculo cardíaco morre ou é danificado. Pacientes diabéticos, com hipertensão arterial, obesos e que sofrem de estressem e depressão, assim como quem tem arritmia cardíaca, completam o grupo de risco da doença.

Não são todas as pessoas que sofrem um infarto que chegam ao óbito. Para aumentar as chances de sobrevivência, é indispensável o atendimento médico com rapidez.

### Sintomas
Dor no peito é o principal sintoma do infarto. O paciente também pode sentir dores estendidas para braços, ombros e pescoço. Na mulher, os sintomas podem ser diferentes: respiração curta, dor na mandíbula, náusea, dor de estômago e sensação de desconforto no peito.

Esses sinais duram aproximadamente 20 minutos e podem ir e voltar. Pacientes diabéticos geralmente não apresentam sintomas, mas também podem ser acometidos pela doença.

#### Outros sintomas são:

Desmaio
Tontura
Falta de ar (principalmente em idosos)
Excesso de suor
Formigamento
O infarto é considerado uma emergência médica. Ao sentir os sintomas, ligue imediatamente para o 192, Serviço de Atendimento Móvel de Urgência (SAMU). As duas primeiras horas são decisivas para o risco de morte.

### 0.1 Loading Data

### Features: 

- Age : Idade do paciente

- Sex : Sexo do paciente
    * 0: Mulher
    * 1: Homem

- exang: Angina induzida por exercício
    * 0: No
    * 1: Yes
    
- ca: Número de vasos no coração (0-4)

- cp : Tipo de dor no peito
    * 1: Angina tipica
    * 2: Angina atipica
    * 3: Não dor - Angina
    * 4: Assintomático
    
- trtbps : Pressão sanguínea em repouso (mm/hg)

- chol : Colesterol em mg/dl

- fbs : Açúcar no sangue em jejum
    * 0: fbs < 120 mg/dl
    * 1: fbs > 120 mg/dl
    
- restecg : Resultados eletrocardiográficos em repouso
    * 0: normal
    * 1: Possui onda ST-T anormal
    * 2: mostrando provável ou definitiva hipertrofia ventricular esquerda
    
- thalachh : Frequência cardíaca máxima alcançada

- oldpeak : Depressão de ST induzida por exercício em relação ao repouso

- slp : A inclinação do segmento ST de pico do exercício
    * 1: inclinação para cima
    * 2: plano
    * 3: inclinação para baixo

- thall : Talassemia
    * 0, 1: normal
    * 2: defeito corrigido
    * 3: defeito reversível

- output : Chance de ter ataque cardíaco
    * 0: Alta
    * 1: Baixa

### Validação Cruzada(Cross-Validation)

Para o resultado ser fiel a realidade e não ficar refém da sorte de uma aleatóriedade, utilizaremos validação cruzada. Para saber mais como funciona essa técnica, leia [este artigo](https://minerandodados.com.br/validacao-cruzada-aprenda-de-forma-simples-como-usar-essa-tecnica/#:~:text=A%20valida%C3%A7%C3%A3o%20cruzada%20%C3%A9%20uma,da%20modelagem%20%C3%A9%20a%20predi%C3%A7%C3%A3o.).

Nesse projeto iremos utilizar o KFold com o numero de folders igual a 10.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
models= []

models.append(('Logistic', LogisticRegression()))
models.append(('DecisionTree', DecisionTreeClassifier()))
models.append(('KNeighbors', KNeighborsClassifier()))
models.append(('SVC', SVC()))
models.append(('AdaBoost', AdaBoostClassifier()))
models.append(('RandomForest', RandomForestClassifier()))
models
```

```python
np.random.seed(seed)

names= []
results= []

for name, model in models:
    kfold= KFold(n_splits= 10, shuffle= True)
    cv= cross_val_score(estimator= model, X= x_train_scaler, y= y_train, scoring= 'accuracy', cv= kfold)*100
    media= cv.mean()
    std= cv.std()
    print('{:15} : Acurácia Média {:.2f}% - Desvio Padrão {:.2f}%'.format(name, media, std))
    names.append(name)
    results.append(cv)
```

```python
plt.figure(figsize= (10, 5))
plt.boxplot(results, labels= names)
plt.title('Comparação dos modelos', fontsize= 20)
plt.show()
```

- A partir da Validação Cruzada, comprovamos que o RandomForestClassifier foi o modelo com a melhor média e o terceiro melhor desvio padrão

## 8.0. HYPERPARAMETER FINE TUNNING
Sendo o modelo RandomForestClassifier como o mais prolífico, vamos analisá-lo utilizando o GridSearchCV para encontrar o melhor parâmetro.

>* O GridSearchCV é uma ferramenta usada para automatizar o processo de ajuste dos parâmetros de um algoritmo, pois ele fará de maneira sistemática diversas combinações dos parâmetros e depois de avaliá-los os armazenará num único objeto.
Para saber mais, leia [este artigo](https://andersonuyekita.github.io/notebooks/blog/2019/03/21/como-usar-o-gridsearchcv/#:~:text=O%20GridSearchCV%20%C3%A9%20uma%20ferramenta,os%20armazenar%C3%A1%20num%20%C3%BAnico%20objeto.).

### 8.1. GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Semente global
np.random.seed(seed)

n_estimators_values= [50, 100, 150, 200]
criterion_values= ['gini', 'entropy']
max_depth_values= [2, 3, 4, 5]
params_values= dict(n_estimators= n_estimators_values, criterion= criterion_values, max_depth= max_depth_values)


modelo= RandomForestClassifier()
kfold= KFold(n_splits= 10, shuffle= True)
grid= GridSearchCV(estimator= modelo, param_grid= params_values, scoring= 'accuracy', cv= kfold)
grid_result= grid.fit(x_train_scaler, y_train)
print('Melhor: {:.2f}% com {}\n'.format(grid_result.best_score_*100, grid_result.best_params_))

means= grid_result.cv_results_['mean_test_score']
stds= grid_result.cv_results_['std_test_score']
params= grid_result.cv_results_['params']
for mean, std, param in zip(means, stds, params):
    print('Acurácia Média {:.2f}% - Desvio Padrão {:.2f}% : Parâmetros {}'.format(mean*100, std*100, param))
```

### 8.2. Treinando e prevendo Random Forest com os melhores parâmetros.

```python
from sklearn.metrics import accuracy_score, classification_report

np.random.seed(seed)

modelo= RandomForestClassifier(criterion= 'entropy', max_depth= 2, n_estimators= 50)

modelo.fit(x_train_scaler, y_train)
y_pred= modelo.predict(x_test_scaler)
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred, digits=4))

print("Accuracy: {:.2f}%\n".format(accuracy_score(y_test, y_pred)*100))
```
